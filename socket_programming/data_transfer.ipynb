```python
import pickle
x = pickle.dumps([99, 100])    # on sending end... convert to byte strings

print(x)
```

```python
pickle.loads(x)   # on receiving end... convert back to object
```

```python
import struct
s = struct.pack('>I', 1234)   # convert simpler types for transmission
print(s)
```

```python
struct.unpack('>I', x)
```
